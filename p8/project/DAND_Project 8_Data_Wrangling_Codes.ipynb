```python
import pandas as pd
import os
from collections import Counter
file = '/Users/JoJo/Documents/Carrier_DB1B_MARKET.csv'
```

## Carrier_Info.csv##

```python
df = pd.read_csv(file)
#print(os.getcwd())
#print(os.listdir(os.getcwd()))
```

```python
df.head()
```

```python
df.columns = map(str.lower, df.columns)
```

```python
df = df[(df['origin'] == 'ORD') & (df['ticket_carrier'] != '99')]
df.head()
```

```python
Counter(df.ticket_carrier)
```

```python
df = df[(df['ticket_carrier'] != '--')]
Counter(df.ticket_carrier)
```

```python
df_clean = df.replace(['AA', 'AS', 'B6', 'BA', 'CO', 'DL', 'IB', 'JL', 'MX', 'NW', 'SN', 'UA', 'US'], 
                      ['American Airlines', 'Alaska Airlines', 'JetBlue Airways', 'British Airways', 'Continental Air Lines',
                       'Delta Air Lines', 'Iberia Air Lines Of Spain', 'Japan Air Lines', 'Compania Mexicana De Aviaci',
                      'Northwest Airlines', 'Brussels Airlines N.V.', 'United Air Lines', 'US Airways'])
```

```python
df_clean.head()
```

```python
df_clean.drop(['operating_carrier'], axis = 1)
```

```python
df_clean.to_csv('Carrier_Info.csv', sep='\t')
```
